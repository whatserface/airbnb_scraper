```python
def get_json(ID):
    url = f"https://www.airbnb.com/rooms/{ID}"
    driver = webdriver.Chrome("/Users/pavelrezabek/Desktop/coding/airbnb/chromedriver")
    driver.get(url)
    html_code = driver.page_source
    soup = BeautifulSoup(html_code)
    string_soup = str(soup)
    jsn = re.search("<script data-deferred-state=\"true\" id=\"data-deferred-state\" type=\"application\\/json\">(.*})<\\/script>", string_soup).group(1)
    jsn = json.loads(jsn)
    return jsn
my_json = get_json(1016153)#https://www.airbnb.com/rooms/1016153 kings - 18081993
```

```python
def get_all_amenities(my_json):
    my_df = pd.json_normalize(my_json["niobeMinimalClientData"][0][1]['data']['presentation']['stayProductDetailPage']['sections']['sections'])
    nrow = pd.json_normalize(my_json["niobeMinimalClientData"][0][1]['data']['presentation']['stayProductDetailPage']['sections']['sections'][my_df.index[my_df["sectionComponentType"] == 'AMENITIES_DEFAULT'].tolist()[0]]['section']["seeAllAmenitiesGroups"]).shape[0]
    
    
    list_of_dfs = []
    for i in list(range(0, nrow)):
        list_of_dfs.append(pd.json_normalize(my_json["niobeMinimalClientData"][0][1]['data']['presentation']['stayProductDetailPage']['sections']['sections'][my_df.index[my_df["sectionComponentType"] == 'AMENITIES_DEFAULT'].tolist()[0]]['section']["seeAllAmenitiesGroups"][i]["amenities"]))
    output_df = pd.concat(list_of_dfs)
    
    return output_df#output_df
    
    
x = get_all_amenities(my_json)
x
```

```python
def get_coordinates(my_json): # nevím jak se zbavit druhého řádku :( 
    my_df = pd.json_normalize(my_json["niobeMinimalClientData"][0][1]['data']['presentation']['stayProductDetailPage']['sections']['sections'])
    to_lookup = "LOCATION_PDP"
    lat = (my_json["niobeMinimalClientData"][0][1]['data']['presentation']['stayProductDetailPage']['sections']['sections'][my_df.index[my_df["sectionComponentType"] == to_lookup].tolist()[0]]["section"])['lat']
    long = (my_json["niobeMinimalClientData"][0][1]['data']['presentation']['stayProductDetailPage']['sections']['sections'][my_df.index[my_df["sectionComponentType"] == to_lookup].tolist()[0]]["section"])['lng']
    to_df = {'lat': [lat,0],'long':[long,0]}
    df = pd.DataFrame(data=to_df)
    return df
get_coordinates(my_json)
```

```python
def get_size(my_json):
    my_df = pd.json_normalize(my_json["niobeMinimalClientData"][0][1]['data']['presentation']['stayProductDetailPage']['sections']['sections'])
    to_lookup = "OVERVIEW_DEFAULT"
    my_list = (my_json["niobeMinimalClientData"][0][1]['data']['presentation']['stayProductDetailPage']['sections']['sections'][my_df.index[my_df["sectionComponentType"] == to_lookup].tolist()[0]]["section"]["detailItems"])
    guests = re.findall(r'\d+', my_list[0]["title"])[0]
    bedrooms = re.findall(r'\d+', my_list[1]["title"])[0]
    beds = re.findall(r'\d+', my_list[2]["title"])[0]
    baths = re.findall(r'\d+', my_list[3]["title"])[0]
    to_df = {'guests': [guests,0],'bedrooms':[bedrooms,0],'beds': [beds,0],'baths':[baths,0]}
    df = pd.DataFrame(data=to_df)
    return df
get_size(my_json)
```
